```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flight = sns.load_dataset('flights')
```

```python
flight.head()
```

```python
#first to convert the data into matrix then plot  by using pivot method
```

```python
flight2 = flight.pivot_table(index='month', columns='year', values='passengers')
```

```python
flight2
```

```python
tips2 = tips.pivot_table(index='day', columns='tip', values='total_bill')
```

```python
tips2
```

```python
#now plot
```

```python
sns.heatmap(flight2,lw=4,cmap='cool')
```
